```python
player_dict = {}

for player in df_list:
    if player in player_dict:
        player_dict[player] += 1   # increase count if already exists
    else:
        player_dict[player] = 1    # add player with count = 1

max_player = max(player_dict, key=player_dict.get)

print(max_player, player_dict[max_player])

#print(player_dict[max_value], max_value)
```

DataFrame Functions

```python
# | Operation      | Works on     | Main purpose                      |
# |----------------|--------------|-----------------------------------|
# | value_counts   | Series/Index | Count unique values               |
# | sort_values    | Series/DF    | Sort by values                    |
# | sort_index     | Series/DF    | Sort by index labels              |
# | set_index      | DataFrame    | Set one/more cols as index        |
# | reset_index    | DataFrame    | Turn index into column            |
# | unique/nunique | Series       | Unique values / count             |
# | isnull         | Series/DF    | Missing value mask                |
# | dropna         | Series/DF    | Remove missing data               |
# | fillna         | Series/DF    | Fill missing data                 |
# | drop_duplicates| Series/DF    | Remove duplicate observations     |
# | drop           | Series/DF    | Remove rows/columns by label      |
# | apply          | Series/DF    | Apply custom function             |
# | isin           | Series/DF    | Membership test                   |
# | corr           | Series/DF    | Correlation                       |
# | nlargest/smallest| DataFrame  | Top/bottom N rows by column       |
# | insert         | DataFrame    | Insert column at position         |
# | copy           | Series/DF    | Safe copy for further changes     |
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```
